```python
# ============================================================
# Malaria detection (Parasitized vs Uninfected) using CNN+ViT
# End-to-end: subset creation, data pipeline, model, training,
# evaluation, Grad-CAM visualization (explainability).
# ------------------------------------------------------------
# Colab-ready. Adjust DATASET_PATH to your Drive location.
# ============================================================

# 1. Mount Google Drive (skip if not using Colab)
try:
    from google.colab import drive
    drive.mount('/content/drive')
    IN_COLAB = True
except:
    IN_COLAB = False

# 2. Paths and setup
import os, random, math, shutil, pathlib

# Adjust to your dataset path (Kaggle/NIH malaria dataset structure)
DATASET_PATH = "/content/drive/MyDrive/Research/archive/cell_images/cell_images"  # contains "Parasitized" and "Uninfected"
WORK_DIR = "/content/malaria_subset_4k"  # will be created
os.makedirs(WORK_DIR, exist_ok=True)

classes = ["Parasitized", "Uninfected"]

# 3. Create 2000-image subset (1000 per class) and split 70/15/15
files_by_class = {}
SAMPLES_PER_CLASS = 1000
for cls in classes:
    cls_dir = pathlib.Path(DATASET_PATH) / cls
    files = [str(p) for p in cls_dir.glob("*") if p.suffix.lower() in [".png", ".jpg", ".jpeg"]]
    random.shuffle(files)
    if len(files) < SAMPLES_PER_CLASS:
        raise ValueError(f"Not enough images in {cls}: found {len(files)}, need {SAMPLES_PER_CLASS}")
    files_by_class[cls] = files[:SAMPLES_PER_CLASS]

splits = {"train": 0.7, "val": 0.15, "test": 0.15}
for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(WORK_DIR, split, cls), exist_ok=True)

for cls in classes:
    files = files_by_class[cls]
    n = len(files)
    n_train = math.floor(n * splits["train"])     # 700 per class
    n_val   = math.floor(n * splits["val"])       # 150 per class
    train_files = files[:n_train]
    val_files   = files[n_train:n_train+n_val]
    test_files  = files[n_train+n_val:]

    for f in train_files:
        shutil.copy(f, os.path.join(WORK_DIR, "train", cls, os.path.basename(f)))
    for f in val_files:
        shutil.copy(f, os.path.join(WORK_DIR, "val", cls, os.path.basename(f)))
    for f in test_files:
        shutil.copy(f, os.path.join(WORK_DIR, "test", cls, os.path.basename(f)))

print("Subset created at:", WORK_DIR)

# 4. Imports and data pipeline
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

AUTOTUNE = tf.data.AUTOTUNE
BATCH_SIZE = 32
IMAGE_SIZE = (224, 224)

# Data augmentation layer (applied on the fly)
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.1),
    layers.RandomBrightness(0.1),
    layers.RandomContrast(0.1),
], name="augmentation")

def make_dataset(split_dir, shuffle, augment=False):
    ds = tf.keras.preprocessing.image_dataset_from_directory(
        split_dir,
        labels="inferred",
        label_mode="binary",
        image_size=IMAGE_SIZE,
        interpolation="bilinear",
        batch_size=BATCH_SIZE,
        shuffle=shuffle
    )
    # Normalize to [0,1] then standardize with ImageNet mean/std (optional but helpful)
    def preprocess(x, y):
        x = tf.cast(x, tf.float32) / 255.0
        mean = tf.constant([0.485, 0.456, 0.406], shape=(1,1,1,3))
        std  = tf.constant([0.229, 0.224, 0.225], shape=(1,1,1,3))
        x = (x - mean) / std
        return x, y
    ds = ds.map(preprocess, num_parallel_calls=AUTOTUNE)

    if augment:
        ds = ds.map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE)

    return ds.cache().prefetch(AUTOTUNE)

train_ds = make_dataset(os.path.join(WORK_DIR, "train"), shuffle=True, augment=True)
val_ds   = make_dataset(os.path.join(WORK_DIR, "val"), shuffle=False, augment=False)
test_ds  = make_dataset(os.path.join(WORK_DIR, "test"), shuffle=False, augment=False)

# 5. CNN + ViT hybrid model
def transformer_block(x, num_heads, mlp_dim, dropout=0.1, name_prefix="blk"):
    # LayerNorm + MHA (return attention scores for explainability)
    x_norm = layers.LayerNormalization(epsilon=1e-6, name=f"{name_prefix}_ln1")(x)
    mha_layer = layers.MultiHeadAttention(
        num_heads=num_heads,
        key_dim=x.shape[-1],
        dropout=dropout,
        output_shape=x.shape[-1],
        name=f"{name_prefix}_mha",
        use_bias=True
    )
    attn_out, attn_scores = mha_layer(x_norm, x_norm, return_attention_scores=True)
    x = layers.Add(name=f"{name_prefix}_mha_res")([x, attn_out])

    # LayerNorm + MLP
    x_norm = layers.LayerNormalization(epsilon=1e-6, name=f"{name_prefix}_ln2")(x)
    mlp = layers.Dense(mlp_dim, activation="gelu", name=f"{name_prefix}_mlp_fc1")(x_norm)
    mlp = layers.Dropout(dropout, name=f"{name_prefix}_mlp_drop")(mlp)
    mlp = layers.Dense(x.shape[-1], name=f"{name_prefix}_mlp_fc2")(mlp)
    x = layers.Add(name=f"{name_prefix}_mlp_res")([x, mlp])
    return x, attn_scores

def build_cnn_vit(input_shape=(224, 224, 3),
                  cnn_channels=(32, 64, 128),
                  patch_size=4, # Changed from 16 to 4
                  embed_dim=256,
                  num_transformer_blocks=4,
                  num_heads=8,
                  mlp_dim=512,
                  dropout=0.1):
    inputs = keras.Input(shape=input_shape, name="input")

    # CNN stem: downsample and enrich local features
    x = inputs
    for i, c in enumerate(cnn_channels):
        x = layers.Conv2D(c, 3, padding="same", activation="relu", name=f"conv_{i}_a")(x)
        x = layers.Conv2D(c, 3, padding="same", activation="relu", name=f"conv_{i}_b")(x)
        x = layers.MaxPool2D(name=f"pool_{i}")(x)  # reduces H,W

    # Ensure spatial dims divisible by patch_size
    h, w = x.shape[1], x.shape[2]
    target_h = (h // patch_size) * patch_size
    target_w = (w // patch_size) * patch_size
    x = layers.Resizing(target_h, target_w, interpolation="bilinear", name="resize_tokens")(x)

    # Linear projection of patches (Conv with stride=patch_size acts as patchify+proj)
    x = layers.Conv2D(embed_dim, kernel_size=patch_size, strides=patch_size, padding="valid", name="patch_proj")(x)  # [B, H', W', D]
    h_tokens, w_tokens = x.shape[1], x.shape[2]
    num_tokens = h_tokens * w_tokens

    # Flatten to tokens
    x = layers.Reshape((num_tokens, embed_dim), name="flatten_tokens")(x)

    # Positional embedding
    positions = tf.range(start=0, limit=num_tokens, delta=1)
    pos_embed_layer = layers.Embedding(input_dim=num_tokens, output_dim=embed_dim, name="pos_embedding")
    pos_embed = pos_embed_layer(positions)
    x = layers.Add(name="add_pos")([x, pos_embed])

    # Transformer encoder (collect last attention scores for visualization)
    attn_scores_list = []
    for b in range(num_transformer_blocks):
        x, attn_scores = transformer_block(x, num_heads=num_heads, mlp_dim=mlp_dim, dropout=dropout, name_prefix=f"blk{b}")
        attn_scores_list.append(attn_scores)

    # Token pooling
    x = layers.LayerNormalization(epsilon=1e-6, name="pre_pool_ln")(x)
    x = layers.GlobalAveragePooling1D(name="token_pool")(x)

    # Classification head
    x = layers.Dropout(dropout, name="head_drop1")(x)
    x = layers.Dense(mlp_dim, activation="gelu", name="head_fc1")(x)
    x = layers.Dropout(dropout, name="head_drop2")(x)
    outputs = layers.Dense(1, activation="sigmoid", name="pred")(x)

    # For explainability, also output last block's attention scores (optional)
    model = keras.Model(inputs, outputs, name="cnn_vit_hybrid")
    return model

model = build_cnn_vit(
    input_shape=(224, 224, 3),
    cnn_channels=(32, 64, 128),
    patch_size=4, # Changed from 16 to 4
    embed_dim=256,
    num_transformer_blocks=4,
    num_heads=8,
    mlp_dim=512,
    dropout=0.1
)

# Compile
model.compile(
    optimizer=keras.optimizers.Adam(3e-4),
    loss="binary_crossentropy",
    metrics=[
        keras.metrics.BinaryAccuracy(name="acc"),
        keras.metrics.AUC(name="auc"),
        keras.metrics.Precision(name="precision"),
        keras.metrics.Recall(name="recall")
    ]
)

model.summary()

# 6. Train
callbacks = [
    keras.callbacks.EarlyStopping(patience=6, restore_best_weights=True, monitor="val_acc"),
    keras.callbacks.ReduceLROnPlateau(patience=3, factor=0.5, monitor="val_loss")
]

history = model.fit(train_ds, validation_data=val_ds, epochs=20, callbacks=callbacks)

# 7. Evaluation
test_metrics = model.evaluate(test_ds)
print("Test metrics:", dict(zip(model.metrics_names, test_metrics)))

# Predict and compute thresholded accuracy, ROC, PR, confusion matrix, classification report
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve

# Collect y_true and y_pred
y_true_list, y_pred_list = [], []
for x_batch, y_batch in test_ds:
    y_true_list.append(y_batch.numpy())
    y_pred_list.append(model.predict(x_batch, verbose=0))
y_true = np.concatenate(y_true_list, axis=0).ravel().astype(int)
y_pred = np.concatenate(y_pred_list, axis=0).ravel()
y_hat = (y_pred >= 0.5).astype(int)

acc = (y_hat == y_true).mean()
print(f"Thresholded test accuracy: {acc:.4f}")

print("\n=== Classification Report ===")
print(classification_report(y_true, y_hat, target_names=classes))

cm = confusion_matrix(y_true, y_hat)
print("Confusion matrix:\n", cm)

# Plot curves and confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()

# Training curves
history_df = pd.DataFrame(history.history)
plt.figure()
history_df[['acc', 'val_acc']].plot()
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure()
history_df[['loss', 'val_loss']].plot()
plt.title("Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()

# ROC curve
fpr, tpr, _ = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.3f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Precision-Recall curve
prec, rec, _ = precision_recall_curve(y_true, y_pred)
plt.figure()
plt.plot(rec, prec)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.tight_layout()
plt.show()

# 8. Save model
save_path = "/content/malaria_model_cnn_vit.keras" if IN_COLAB else "./malaria_model_cnn_vit.keras"
model.save(save_path)
print("Model saved to:", save_path)

# 9. Explainability: Grad-CAM for CNN part (last Conv2D layer)
import cv2

def get_last_conv_layer(model):
    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name
    raise ValueError("No Conv2D layer found in the model.")

def compute_gradcam(model, img_tensor):
    last_conv_name = get_last_conv_layer(model)

    grad_model = tf.keras.models.Model(
        inputs=model.input,
        outputs=[
            model.get_layer(last_conv_name).output,
            model.output
        ]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_tensor, training=False)
        loss = predictions[:, 0]  # binary target score

    grads = tape.gradient(loss, conv_outputs)
    # Global average pooling of gradients across spatial dims
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]

    # Weight conv feature maps by pooled grads
    cam = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)  # [H, W]
    cam = tf.nn.relu(cam)
    cam = cam / (tf.reduce_max(cam) + 1e-8)

    cam = cv2.resize(cam.numpy(), IMAGE_SIZE)
    return cam

def overlay_gradcam(original_batch_image, cam, alpha=0.5):
    # original_batch_image is a single image tensor from dataset after normalization
    # Convert back to displayable RGB [0,255]
    img = original_batch_image.numpy()
    # undo normalization (approximate): clip after re-scaling; display purpose only
    img = (img - img.min()) / (img.max() - img.min() + 1e-8)
    img = (img * 255).astype("uint8")

    heatmap = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    overlay = cv2.addWeighted(img, 1 - alpha, heatmap, alpha, 0)
    return overlay

# Take one batch and visualize Grad-CAM on a few samples
for images, labels in test_ds.take(1):
    # images are normalized; use as input to model
    # Select first image
    img_tensor = images[0:1]  # [1, H, W, 3]
    cam = compute_gradcam(model, img_tensor)
    overlay = overlay_gradcam(images[0], cam)

    plt.figure(figsize=(6,6))
    plt.imshow(overlay)
    plt.axis("off")
    plt.title("Grad-CAM (CNN part of CNN+ViT)")
    plt.tight_layout()
    plt.show()

# 10. Optional: Visualize token attention (simplified)
# Note: Our transformer_block uses Keras MultiHeadAttention, but we didn't expose scores in the final model outputs.
# Below is a small utility that builds a submodel to fetch attention scores of a chosen block for visualization.

def build_attention_extractor(full_model, block_index=0):
    # Find names for the desired block's LayerNormalization before MHA
    ln1_name = f"blk{block_index}_ln1"
    mha_name = f"blk{block_index}_mha"

    # Build a model that outputs the attention scores by tapping into the MHA layer
    # We need to recreate the computation graph up to the MHA call.
    # Easier approach: rebuild a tiny model mirroring the inputs-to-MHA path.
    # Since Keras layers are named, we can get intermediate outputs via the functional API.

    # Get intermediate layer outputs
    input_tensor = full_model.get_layer("input").output
    # Traverse up to "flatten_tokens"
    x = full_model.get_layer("conv_0_a")(input_tensor)
    x = full_model.get_layer("conv_0_b")(x)
    x = full_model.get_layer("pool_0")(x)

    x = full_model.get_layer("conv_1_a")(x)
    x = full_model.get_layer("conv_1_b")(x)
    x = full_model.get_layer("pool_1")(x)

    x = full_model.get_layer("conv_2_a")(x)
    x = full_model.get_layer("conv_2_b")(x)
    x = full_model.get_layer("pool_2")(x)

    x = full_model.get_layer("resize_tokens")(x)
    x = full_model.get_layer("patch_proj")(x)
    x = full_model.get_layer("flatten_tokens")(x)

    pos_add = full_model.get_layer("add_pos")
    # Recreate positions addition by calling the layer
    # The layer "add_pos" took tokens and pos_embed; we can re-use it by getting its inbound nodes
    # Simpler: re-fetch embedding and manually add
    num_tokens = x.shape[1]
    positions = tf.range(start=0, limit=num_tokens, delta=1)
    pos_embedding = full_model.get_layer("pos_embedding")(positions)
    x = layers.Add()([x, pos_embedding])

    # LayerNorm before MHA
    ln1 = full_model.get_layer(ln1_name)
    x_norm = ln1(x)

    # MultiHeadAttention layer with return_attention_scores=True
    mha = full_model.get_layer(mha_name)
    # Call MHA explicitly to get attention scores
    attn_out, attn_scores = mha(x_norm, x_norm, return_attention_scores=True)

    extractor = keras.Model(inputs=full_model.input, outputs=attn_scores, name=f"attn_extractor_blk{block_index}")
    return extractor

try:
    attn_extractor = build_attention_extractor(model, block_index=0)
    # Get attention for one image
    for images, labels in test_ds.take(1):
        attn_scores = attn_extractor.predict(images[0:1], verbose=0)  # shape [batch, num_heads, tokens, tokens]
        attn_scores = attn_scores[0]  # [num_heads, tokens, tokens]
        # Average heads
        attn_mean = attn_scores.mean(axis=0)  # [tokens, tokens]
        # Visualize attention matrix
        plt.figure(figsize=(6,5))
        sns.heatmap(attn_mean, cmap="magma")
        plt.title("ViT Attention (Block 0, averaged over heads)")
        plt.xlabel("Key tokens")
        plt.ylabel("Query tokens")
        plt.tight_layout()
    plt.show()
except Exception as e:
    print("Attention visualization skipped:", e)
    print("Note: Attention visualization depends on Keras graph introspection; Grad-CAM is the primary explainability tool here.")
```
